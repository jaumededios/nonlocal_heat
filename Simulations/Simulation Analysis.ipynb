```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('COMSOL_RESULTS/conductance_parametric_h=400.tsv',sep='\t')
```

```python
res = []
res.append(data.fourier[data.l==5E-8])
ls = [0]
width = data.width[data.l==5E-8]

for l in pd.unique(data.l):
    ls.append(l)
    res.append(data.hydro[data.l==l])

for r,l in zip(res,ls):
    plt.plot(width,r,label='l={}'.format(l))
    
plt.legend()
```

```python
for r in zip(*res):
    plt.plot(ls,r)
```
